# 2.1 — League Current Season Standings

Source: `data/team_stats.parquet`  
Columns: Team | GP | GF | GA | GD | Pts | Pts/G | xPts | xPts/G

```python
import sys
from typing import Final

import pandas as pd

sys.path.insert(0, "../../scripts")
from utils import render_table

SEASON: Final[int] = 2026
```

```python
team_stats = pd.read_parquet("../../data/team_stats.parquet")
```

## Standings — sorted by Pts

```python
df = (
    team_stats[team_stats["season_name"] == SEASON]
    .assign(GD=lambda x: x["goals_for"] - x["goals_against"])
    [["team_name", "count_games", "points", "pts_per_game",
      "goals_for", "goals_per_game", "xg_per_game",
      "goals_against", "goals_against_per_game", "xga_per_game", "GD"]]
    .rename(columns={
        "team_name": "Team",
        "count_games": "GP",
        "points": "Pts",
        "pts_per_game": "Pts/G",
        "goals_for": "GF",
        "goals_per_game": "GF/G",
        "xg_per_game": "xGF/G",
        "goals_against": "GA",
        "goals_against_per_game": "GA/G",
        "xga_per_game": "xGA/G",
    })
    .sort_values(["Pts/G", "Pts"], ascending=False)
    .reset_index(drop=True)
)

fig = render_table(
    df,
    title=f"USL Championship {SEASON} — Standings",
    col_formats={"Pts/G": ".2f", "GF/G": ".2f", "xGF/G": ".2f", "GA/G": ".2f", "xGA/G": ".2f"},
    accent_color="#b28350",
    rank_cols={
        "Pts": "high",
        "Pts/G": "high",
        "GF": "high",
        "GF/G": "high",
        "xGF/G": "high",
        "GA": "low",
        "GA/G": "low",
        "xGA/G": "low",
        "GD": "high",
    },
)
display(fig)
```
